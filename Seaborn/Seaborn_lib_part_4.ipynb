###  What is PairGrid?

    * seaborn.PairGrid is a powerful class for creating a grid of plots to visualize pairwise relationships in a dataset,
        especially when working with multivariate data (data with many numeric columns).

 It gives you fine-grained control over:

    * What kind of plots go on the diagonal, upper, and lower parts of the grid

    * How the grid is styled and organized

####  How is it different from pairplot()?

| Feature                          | `pairplot()` (shortcut) | `PairGrid` (customizable)  |
| -------------------------------- | ----------------------- | -------------------------- |
| Quick plot                       |  Yes                   |  No                       |
| Custom plot types                |  Limited               |  Yes                      |
| Different plots in grid sections |  No                    |  Yes (lower, upper, diag) |
| More control                     |                        |  Full control             |



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create grid
g = sns.PairGrid(iris)

# Map plots
g.map_diag(sns.histplot)             # Diagonal: histograms
g.map_upper(sns.scatterplot)         # Upper triangle: scatter plots
g.map_lower(sns.kdeplot, fill=True)  # Lower triangle: density plots

plt.show()
```

####  What happens here?

    * Diagonal plots  show the distribution of each variable (e.g., histograms)

    * Upper triangle  scatter plots between pairs (e.g., SepalLength vs PetalLength)

    * Lower triangle  KDE (contour plots) to show density of points

```python
g = sns.PairGrid(iris, hue="Species", palette="husl")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.add_legend()
```
