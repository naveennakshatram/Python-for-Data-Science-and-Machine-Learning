# Regression Plots

### 📈 What is a Regression Plot?

    A regression plot shows:

        * A scatter plot of two numeric variables

        * A regression line (best-fit line)

        * Optionally, a confidence interval around the regression line

    This helps you:

        * See correlation or linear trends

        * Detect outliers

        * Understand the strength and direction of a relationship

```python
# Importing tips data set

import pandas as pd

tips = pd.read_csv('./Dataset/tips.csv')

tips.head()
```

#### 🎯 For Groups: Use sns.lmplot()

    * If you want to plot regression for different groups (e.g., by gender), use lmplot():

    * Example
    
        sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips)
        
        This will:

        Split the data by sex

        Plot separate regression lines and confidence bands

```python
# Importing seaborn lib

import seaborn as sns

sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips)
```

#### ✅ Using sns.lmplot() with Markers (Multiple Groups)

    To specify markers per group when using lmplot() (with hue), pass a list of markers:


#### 🧩 Marker Options (Matplotlib styles):

| Marker | Description      |
| ------ | ---------------- |
| `'o'`  | Circle (default) |
| `'^'`  | Triangle up      |
| `'s'`  | Square           |
| `'D'`  | Diamond          |
| `'x'`  | X                |
| `'+'`  | Plus             |
| `'*'`  | Star             |

    

```python
sns.lmplot(
    x='total_bill',
    y='tip',
    hue='sex',
    data=tips,
    markers=['D', 's'],  # one for each group
    palette='Set1'
)
```

#### 🔧 Example: Separate by time (Lunch vs Dinner)

### ✅ Use col in sns.lmplot()

        The col parameter lets you split your data into multiple plots based on a categorical column (like time, sex, or day).

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create regression plots split by 'time' in separate columns
sns.lmplot(
    x='total_bill',
    y='tip',
    data=tips,
    col='time',           # separate plots for Lunch and Dinner
    hue='sex',            # color by gender
    markers=['o', 's'],   # different markers for Male and Female
    palette='Set1',
    height=5,
    aspect=1
)

plt.show()


# This Will:

    # Create two side-by-side plots, one for Lunch, one for Dinner.

    # Show regression lines and scatter plots for both sexes (hue='sex').

    # Use different markers ('o' and 's').

    # Each plot is a small multiple (facet).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create multiple regression plots split by 'day'
sns.lmplot(
    x='total_bill',
    y='tip',
    data=tips,
    col='day',           # Separate plot for each day
    hue='sex',           # Color points by gender
    markers=['o', 's'],  # Different markers for Male/Female
    palette='Set2',      
    aspect=1.2,          # Wider plot (width = aspect * height)
    height=5             # Each plot is 5 inches tall
)

plt.show()
```

# Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn plot style
sns.set_style("darkgrid")  # Options: 'white', 'dark', 'ticks', 'whitegrid', 'darkgrid'

# Plot
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='day',              # Separate plot per day
    hue='sex',              # Color by gender
    markers=['^', 's'],     # Triangle for Male, square for Female
    palette='husl',     # Color theme palette='coolwarm', 'Set1', 'husl', 'dark'
    height=5,
    aspect=1.2,
    line_kws={'linewidth': 2, 'linestyle': '--'}  # Dashed, thicker regression line
)

plt.suptitle("Tip vs Total Bill by Day & Gender", y=1.05)
plt.show()
```

## Color

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Use lmplot with custom color palette
sns.lmplot(
    x='total_bill',
    y='tip',
    data=tips,
    hue='sex',                    # Color lines/points by 'sex'
    palette='dark',               # Try 'Set1', 'coolwarm', 'husl', etc.
    markers=['D', '^'],           # Optional: different markers
    height=5,
    aspect=1.2
)

plt.title("Tip vs Total Bill (Color by Sex)")
plt.show()
```

#### 🔧 Common palette Options

| Palette Name | Description                           | Looks Like |
| ------------ | ------------------------------------- | ---------- |
| `'Set1'`     | Bright, distinct colors               | 🔴 🔵 🟢   |
| `'Set2'`     | Softer tones                          | 🟢 🟡 🟣   |
| `'coolwarm'` | Gradient from blue to red             | 🔵➡️🔴     |
| `'husl'`     | Circular color wheel                  | 🌈         |
| `'muted'`    | Softer than default                   | 🟤 🟡 🔵   |
| `'rocket'`   | Dark-red gradient (good for heatmaps) | 🔴 🔶 🟡   |
| `'flare'`    | Bright-red gradient                   | 🔴🟧🟡     |



```python
import seaborn as sns
import matplotlib.pyplot as plt


# Define your custom color palette using hex codes
custom_palette = {
    'Male': '#3498db',     # Bright blue
    'Female': '#e74c3c'    # Vivid red
}

# Create lmplot with custom colors
sns.lmplot(
    x='total_bill',
    y='tip',
    data=tips,
    hue='sex',                  # Color by 'sex'
    palette=custom_palette,     # Use your hex colors
    markers=['D', '^'],
    height=5,
    aspect=1.2
)

plt.title("Tip vs Total Bill with Custom Hex Colors")
plt.show()
```
