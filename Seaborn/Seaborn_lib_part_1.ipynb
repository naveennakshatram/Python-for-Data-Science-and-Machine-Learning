## âœ… Why Use Seaborn?

    * Cleaner, more stylish default visuals

    * Fewer lines of code for complex plots
    
    * Supports dataframes directly (usually with pandas)

    * Built-in support for statistical plots (histogram, boxplot, violin, regression...)

```python
# Example
import pandas as pd

# loading dataset from csv file
tips = pd.read_csv('./Dataset/tips.csv')

tips
```

```python
import seaborn as sns

sns.displot(tips['total_bill'])
```

### ðŸ”¥ Popular Seaborn Plot Types

| Plot Type    | Function            | Description                            |
| ------------ | ------------------- | -------------------------------------- |
| Line Plot    | `sns.lineplot()`    | Trends over time or order              |
| Scatter Plot | `sns.scatterplot()` | Points & relationships                 |
| Bar Plot     | `sns.barplot()`     | Mean of a categorical variable         |
| Count Plot   | `sns.countplot()`   | Counts of categories                   |
| Box Plot     | `sns.boxplot()`     | Distribution + Outliers                |
| Violin Plot  | `sns.violinplot()`  | Kernel density + box plot              |
| Histogram    | `sns.histplot()`    | Frequency distribution                 |
| KDE Plot     | `sns.kdeplot()`     | Smoothed density                       |
| Heatmap      | `sns.heatmap()`     | Color-coded matrix (correlation etc.)  |
| Pair Plot    | `sns.pairplot()`    | Grid of plots for each pair of columns |
| Cat Plot     | `sns.catplot()`     | Combines several categorical plots     |
| Joint Plot   | `sns.jointplot()`   | 2D scatter + histograms or KDE         |


```python
sns.catplot(tips['total_bill'])
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```
