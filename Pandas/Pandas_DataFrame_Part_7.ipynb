# Pandas Built - in - Data Visualisation

```python
# Importing pyplot
import matplotlib.pyplot as plt

# Importing data 
import pandas as pd

tips = pd.read_csv('../Seaborn/Dataset/tips.csv')

tips.head()
```

## Example's

### 1. Line Plot

    Plotting total_bill over the index (just to show usage).

```python
tips['total_bill'].plot()
plt.title("Line Plot of Total Bill")
```

### 2. Bar Plot

    Plot total_bill for each row as bars.

```python
tips['total_bill'].plot.bar()
plt.title("Bar Plot of Total Bill")
```

### 3. Horizontal Bar Plot

```python
tips['total_bill'].plot.barh()
plt.title("Horizontal Bar Plot of Total Bill")
plt.show()
```

### 4. Histogram
Histogram of tip values.

```python
tips['tip'].plot.hist(bins=5, alpha=0.7)
plt.title("Histogram of Tips")
plt.show()
```

### 5. Box Plot
Box plot of total_bill and tip.

```python
tips[['total_bill', 'tip']].plot.box()
plt.title("Box Plot of Total Bill and Tip")
plt.show()
```

### 6. Area Plot

```python
tips.plot.area(alpha=0.4)
plt.title("Area Plot: Total Bill and Tip")
plt.xlabel("Index")
plt.ylabel("Amount ($)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(tips))
plt.fill_between(x, tips['total_bill'], alpha=0.5, label='Total Bill')
plt.fill_between(x, tips['tip'], alpha=0.5, label='Tip')
plt.legend()
plt.title("Area Plot (Matplotlib fill_between)")
plt.xlabel("Index")
plt.ylabel("Amount ($)")
plt.show()
```

# Plotly and Cufflinks

    * Plotly is a popular interactive plotting library.

    * Cufflinks connects Pandas with Plotly, letting you create interactive charts directly from DataFrames.

#### Setup
    
    pip install plotly cufflinks

#### [plotly](https://dash.plotly.com/)

#### [cufflinks](https://github.com/santosjorge/cufflinks)


```python
import plotly.express as px

# Sample data: unemployment rate by US state (fictitious)
data = {
    'state': ['AL', 'AK', 'AZ', 'AR', 'CA'],
    'unemployment': [5.1, 6.2, 7.0, 6.5, 4.3]
}

import pandas as pd
df = pd.DataFrame(data)

fig = px.choropleth(df, 
                    locations='state',       # column with geographic locations
                    locationmode='USA-states', # interpret locations as US states
                    color='unemployment',    # data to color-code
                    scope="usa",             # map scope
                    color_continuous_scale='Viridis',
                    title='US Unemployment Rate by State')

fig.show()
```
