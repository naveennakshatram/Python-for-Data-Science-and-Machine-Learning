```python
# To find the average BasePay.

salaries['BasePay'].mean()

# This will return error because som rows are having data which is not int/float type .
```

```python
# Solution to find average of BasePay is.

# First we need to convert evenry thing into numeric type.

basepay = pd.to_numeric(salaries['BasePay'], errors='coerce')

print(basepay)

print("Average BasePay = ",basepay.mean())
```

```python
print("Highest BasePay = ",basepay.max())
```

```python
# Find the Highest pay for Overtime

print(salaries['OvertimePay'])

# Convert all colums into numaric type

overtime_pay = pd.to_numeric(salaries['OvertimePay'],errors='coerce')

print("Highest Pay of Overtime is ",overtime_pay.max())
```

```python
# Find the Job Title with Employee Name 
# Employee Name is : 'JOSEPH DRISCOLL'

salaries[salaries['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

```python
# Print the TotalPayBenfits of given employee
# Employee Name : 'JOSEPH DRISCOLL'

salaries[salaries['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
# Find the Employee Name who is paid highest Total Pay Benfits

salaries[salaries['TotalPayBenefits'] == salaries['TotalPayBenefits'].max()]['EmployeeName']
```

```python
# Find the Employee Name who is paid Lowest Total Pay Benfits

salaries[salaries['TotalPayBenefits'] == salaries['TotalPayBenefits'].min()]['EmployeeName']
```

```python
# Find howmany unique job titles ..?

salaries['JobTitle'].nunique()
```

```python
# What are the top 5 most common Jobs ?

salaries['JobTitle'].value_counts().head()
```

```python
# How many job titles were represented by only one person in 2013?

sum(salaries[salaries['Year'] == 2013]['JobTitle'].value_counts() == 1)
```

```python
# How many people have the word cheif in their job Title?

# using custom function we are solving this...!


def checkCHIEF(value):
    if 'CHIEF' in value or 'chief' in value:
        return True
    return False

booleans = salaries['JobTitle'].apply(lambda title : checkCHIEF(title))

len(salaries[booleans])
```
