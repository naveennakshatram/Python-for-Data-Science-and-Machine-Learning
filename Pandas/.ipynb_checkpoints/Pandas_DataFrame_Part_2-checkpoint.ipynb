# Operators on DataFrame

### 📣 Comparison Operators 

| Operator | Meaning                  | Example  | Result |
| -------- | ------------------------ | -------- | ------ |
| `>`      | Greater than             | `5 > 3`  | `True` |
| `<`      | Less than                | `2 < 4`  | `True` |
| `>=`     | Greater than or equal to | `5 >= 5` | `True` |
| `<=`     | Less than or equal to    | `3 <= 4` | `True` |
| `==`     | Equal to                 | `5 == 5` | `True` |
| `!=`     | Not equal to             | `5 != 3` | `True` |


```python
import pandas as pd

from numpy.random import randn

df = pd.DataFrame(data=randn(5,5), index=['A','B','C','D','E'], columns=['Z','Y','X','W','V'])

df
```

```python
df > 0 

# Applied > 0 on complete DataFrame (all rows and cols) 
# This will return only boolean value.
```

```python
df[df > 0]

# the boolean response we are giving DataFrame Object so we can see actual values of DataFrame

# NaN means False value.
```

```python
df['W'] > 0   

# Applied > 0 on  'W' column
```

```python
df[df['W']>0]

# Giving the boolean response to DataFrame object to get actual values.

##  Note :  Just observe column 'W' only.
```

```python
# To read only 'W' column (Which are > 0).

df[df['W']>0]['W']
```

```python
df
```

```python
# Reding only 'D' Row

df.loc['D']
```

```python
# Reding only 'D' Row > 0

df.loc['D'] > 0
```

```python
# Reding only 'D' Row > 0 values

df[df.loc['D'] > 0]
```

## Now, the problem with above code:

#### df[df.loc['D'] > 0]

* We trying to filter rows of df using this boolean Series.

* But the boolean Series here has 5 elements (one per column), while filtering df expects a boolean array of length 5 (rows).

* The boolean array length matches the columns count, but filtering rows needs a boolean array matching the rows count.

* That mismatch causes an error.

## Fix

```python
df.loc[:, df.loc['D'] > 0]  # Using Slice on row...!

# observe only row 'D' values.
```

### ✳️ Explanation of df.loc[] parameters

#### The syntax is:  df.loc[row_indexer, column_indexer]

* row_indexer: specifies which rows to select (by label)

* column_indexer: specifies which columns to select (by label)

| Usage                        | Meaning                           | Returns        |
| ---------------------------- | --------------------------------- | -------------- |
| `df.loc['A']`                | Row with index label 'A'          | Series (1 row) |
| `df.loc[['A','B']]`          | Rows with labels A and B          | DataFrame      |
| `df.loc['B':'D']`            | Rows from B to D inclusive        | DataFrame      |
| `df.loc[:, 'Z']`             | All rows, column 'Z'              | Series         |
| `df.loc[:, ['Z','Y']]`       | All rows, columns Z and Y         | DataFrame      |
| `df.loc['B':'E', 'Y':'W']`   | Rows B to E, columns Y to W       | DataFrame      |
| `df.loc[df['Z'] > 0]`        | Rows where column Z > 0           | DataFrame      |
| `df.loc[:, df.loc['D'] > 0]` | All rows, columns where row D > 0 | DataFrame      |


```python
df.loc['D':, df.loc['D'] > 0]
```

```python
df.loc['D':'D', df.loc['D'] > 0]
```

### DataFrame.set_index(keys, *, drop=True, append=False, inplace=False, verify_integrity=False)

Set the DataFrame index using existing columns.

* Set the DataFrame index (row labels) using one or more existing columns or arrays (of the correct length).
* 
* The index can replace the existing index or expand on it.

```python
df.set_index('names')  # Setting the 'names' colum as a index to DataFrame

# Note this is temporary only it will not effect our original DataFrame
```

```python
df  # original DataFrame.

# if you pass inplace = True it will be Permanent chagne.
```

### DataFrame.reset_index(level=None, *, drop=False, inplace=False, col_level=0, col_fill='', allow_duplicates=<no_default>, names=None)

Reset the index, or a level of it.

* Reset the index of the DataFrame, and use the default one instead.
* 
* If the DataFrame has a MultiIndex, this method can remove one or more levels.
