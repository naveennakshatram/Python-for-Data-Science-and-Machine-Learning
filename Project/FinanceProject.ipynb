### Working with yfinance

[yfinance](https://ranaroussi.github.io/yfinance/)

```python
import yfinance as yf
import datetime

# Define the date range
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)

# # Download Bank of America (BAC) data
# df = yf.download("BAC", start=start, end=end)

# Explicitly set auto_adjust (True or False depending on your use case)
BAC = yf.download("BAC", start=start, end=end, auto_adjust=False)

# Display first few rows
print(BAC.head())
```

```python
# Goldman Sachs

# Explicitly set auto_adjust (True or False depending on your use case)
GS = yf.download("GS", start=start, end=end, auto_adjust=False)

# Display first few rows
print(GS.head())
```

```python
# JPMorgan Chase

# Explicitly set auto_adjust (True or False depending on your use case)
JPM = yf.download("JPM", start=start, end=end, auto_adjust=False)

# Display first few rows
print(JPM.head())
```

```python
# Morgan Stanley

# Explicitly set auto_adjust (True or False depending on your use case)
MS = yf.download("MS", start=start, end=end, auto_adjust=False)

# Display first few rows
print(MS.head())
```

```python
# Wells Fargo

# Explicitly set auto_adjust (True or False depending on your use case)
WFC = yf.download("WFC", start=start, end=end, auto_adjust=False)

# Display first few rows
print(WFC.head())
```

### Create a list of the ticker symobols

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

### Use pd.concat to conactenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list.

```python
import pandas as pd

bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)

bank_stocks.head()
```

### Pandas Dataframe XS

[Documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.xs.html#pandas-dataframe-xs:~:text=pandas.DataFrame.xs-,pandas.DataFrame.xs,-%23)

```python
# Example on xs

d = {'num_legs': [4, 4, 2, 2],
     'num_wings': [0, 0, 2, 2],
     'class': ['mammal', 'mammal', 'mammal', 'bird'],
     'animal': ['cat', 'dog', 'bat', 'penguin'],
     'locomotion': ['walks', 'walks', 'flies', 'walks']}

df = pd.DataFrame(data=d)

df = df.set_index(['class', 'animal', 'locomotion'])

df
```

```python
# Get values at specified index
df.xs('mammal')
```

```python
# Get values at several indexes
df.xs(('mammal', 'dog', 'walks'))
```

```python
# Get values at specified index and level

df.xs('cat', level=1)
```

```python
# Get values at several indexes and levels

df.xs(('bird', 'walks'),level=[0, 'locomotion'])
```

```python
# Get values at specified column and axis

df.xs('num_wings', axis=1)
```

### We can use panads pct_change() method on the close column to create a column representing this return value. Create a for loop that goes and for each Bank stock Ticker creates this returns column and set's it as acolumn in the returns DataFrame.

```python
# Creating an empty return DataFrame

returns = pd.DataFrame()

returns
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```
