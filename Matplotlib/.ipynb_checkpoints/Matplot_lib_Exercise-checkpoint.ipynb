```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Preparing some co-ordinates for plotting

import numpy as np

x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise - 1

```python
# Create a figure object called fig using plt.figure()
# Use add_axes to add an axis to the figure canvas at [0,0,1,1] and assing to a variable.
# Plot(x,y) on the axes and set the labels and titles.

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('Exercise -1')
axes.set_xlabel('X - Co ordinates')
axes.set_ylabel('Y - Co ordinates')
```

## Exercise - 2

```python
# Create a figure object and put two axes on it axes1 and axes2. Located at [0,0,1,1] and [0.2,0.5,.2,.2].

fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
# Now plot(x,y) on both axes and call your figure object to show it.

fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes2.plot(x,y)
```

## Exercise - 3

```python
# Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

# Now use x,y and Z arrays to recreate the plot.
axes1.plot(x,z)
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(x,y)
axes2.set_title('Zoom')
```

# Exercise - 4

```python
# Use plt.subplots(nrows=1,ncols=2) to create the plot.

fig, axes = plt.subplots(nrows=1,ncols=2)

for ax in axes:
    ax.plot(x,y)
```

```python
# Use plt.subplots(nrows=1,ncols=2) to create the plot.

fig, axes = plt.subplots(nrows=1,ncols=2)

# Now plot(x,y) and (x,z) on the axes and change plot design.

# plot - 1
axes[0].plot(x,y,color='red',linewidth=3)

# plot - 2
axes[1].plot(x,z,color='#FF1199',linewidth=2)
```

```python
# Use plt.subplots(nrows=1,ncols=2) to create the plot.

# Now change the width and height of a plot.  figsize=(12,2)  -- 12 is width 2 is height

fig, axes = plt.subplots(1,2,figsize=(14,2))

# Now plot(x,y) and (x,z) on the axes and change plot design.

# plot - 1
axes[0].plot(x,y,color='red',linewidth=3)

# plot - 2
axes[1].plot(x,z,color='#FF1199',linewidth=2)

```
