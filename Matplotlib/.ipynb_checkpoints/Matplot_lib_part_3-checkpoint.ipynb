# Plot Types

## ðŸ”¹ 1. Line Plot
    * Shows trends over time or ordered categories.
    
        plt.plot(x, y)
        
## ðŸ”¹ 2. Scatter Plot
    * Shows relationships or distributions of individual points.

        plt.scatter(x, y)
        
## ðŸ”¹ 3. Bar Plot
    * Used for comparing categories.

    plt.bar(x, y)              # Vertical
    plt.barh(x, y)             # Horizontal
    
## ðŸ”¹ 4. Histogram
    * Shows frequency distribution of numeric data.

    plt.hist(data, bins=10)
    
## ðŸ”¹ 5. Pie Chart
    * Shows proportions of a whole.

    plt.pie(values, labels=labels, autopct='%1.1f%%')
    
## ðŸ”¹ 6. Box Plot (Box-and-Whisker)
    * Shows distribution, median, and outliers.

    plt.boxplot(data)
    
## ðŸ”¹ 7. Area Plot
    * Filled version of a line plot.

    plt.fill_between(x, y)
    
## ðŸ”¹ 8. Stacked Bar Plot
    * Used for showing parts of multiple categories.

    plt.bar(x, y1)
    plt.bar(x, y2, bottom=y1)
    
## ðŸ”¹ 9. Heatmap (via seaborn or imshow)
    * Visualizes 2D data using color intensity.

    plt.imshow(matrix, cmap='hot', interpolation='nearest')
    
## ðŸ”¹ 10. 3D Plot
    * For 3D data visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y = np.array([10, 15, 13, 17, 20])
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 17, 35, 29, 12]

# Create a 5x2 grid of subplots
fig, axs = plt.subplots(5, 2, figsize=(14, 20))
fig.suptitle("Examples of Common Matplotlib Plot Types", fontsize=16)

# 1. Line Plot
axs[0, 0].plot(x, y, color='blue', marker='o')
axs[0, 0].set_title("Line Plot")

# 2. Scatter Plot
axs[0, 1].scatter(x, y, color='green')
axs[0, 1].set_title("Scatter Plot")

# 3. Bar Plot
axs[1, 0].bar(categories, values, color='orange')
axs[1, 0].set_title("Bar Plot")

# 4. Horizontal Bar Plot
axs[1, 1].barh(categories, values, color='purple')
axs[1, 1].set_title("Horizontal Bar Plot")

# 5. Histogram
axs[2, 0].hist(data, bins=20, color='grey')
axs[2, 0].set_title("Histogram")

# 6. Pie Chart
axs[2, 1].pie(values, labels=categories, autopct='%1.1f%%')
axs[2, 1].set_title("Pie Chart")

# 7. Box Plot
axs[3, 0].boxplot([data[:100], data[100:200], data[200:300]])
axs[3, 0].set_title("Box Plot")

# 8. Area Plot
axs[3, 1].fill_between(x, y, color='skyblue', alpha=0.5)
axs[3, 1].set_title("Area Plot")

# 9. Stacked Bar Plot
axs[4, 0].bar(x, y, label='y1')
axs[4, 0].bar(x, y // 2, bottom=y, label='y2')
axs[4, 0].set_title("Stacked Bar Plot")
axs[4, 0].legend()

# 10. Heatmap using imshow
matrix = np.random.rand(5, 5)
im = axs[4, 1].imshow(matrix, cmap='viridis')
axs[4, 1].set_title("Heatmap")
fig.colorbar(im, ax=axs[4, 1])

# Layout adjustment
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
