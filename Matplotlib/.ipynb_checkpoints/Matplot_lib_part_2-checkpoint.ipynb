# Plot Appearance

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

# perparing 2 arrays for ploting

x = np.array([10,20,30,40,50])

y = x * 2
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

#  axes.plot(x,y,color='green')  # Default color name as a prameter value.

axes.plot(x,y,color='#FF55CC') # RGB Color Format.
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#889900',linewidth=5) # Graph Line Width starts from 0.1
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#889900',linewidth=5, alpha=0.5) # alpha for color dark ness and fadeout.
```

## Common linestyle values:

| `linestyle` | Shortcut      | Description   |
| ----------- | ------------- | ------------- |
| `'solid'`   | `'-'`         | Solid line    |
| `'dashed'`  | `'--'`        | Dashed line   |
| `'dashdot'` | `'-.'`        | Dash-dot line |
| `'dotted'`  | `':'`         | Dotted line   |
| `'None'`    | `''` or `' '` | No line       |


```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#889900',linewidth=5,linestyle='--') # line style
```

### we can write linewidth as 'lw' and linestyle as 'ls'

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#889900',lw=2,ls='dashdot')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#889900',lw=2,ls='-', marker="1", markersize=20)
```

### Common marker values:

| `marker`         | Description      |               |
| ---------------- | ---------------- | ------------- |
| `'o'`            | Circle           |               |
| `'s'`            | Square           |               |
| `'^'`            | Triangle up      |               |
| `'v'`            | Triangle down    |               |
| `'>'`            | Triangle right   |               |
| `'<'`            | Triangle left    |               |
| `'D'`            | Diamond          |               |
| `'p'`            | Pentagon         |               |
| `'*'`            | Star             |               |
| `'+'`            | Plus             |               |
| `'x'`            | Cross (x)        |               |
| `'.'`            | Point (tiny dot) |               |
| `','`            | Pixel            |               |
| \`'              | '\`              | Vertical line |
| `'_'`            | Horizontal line  |               |
| `'None'` or `''` | No marker        |               |


```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='blue',lw=2,ls='-', marker="o", markersize=20,markerfacecolor='red', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#889900',lw=2,ls='--')
```
