# Functions

```python
import numpy as np

x = np.ndarray(5)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# A simple plot showing x and y 

plt.plot(x,y)
```

```python
# Showing X and Y Labels with title on Plot

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("Demo Ploting")
```

### Creating Sub-Plots

```python
# Creating multiple plots (Sub Plots)

plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(x,x,'r-')  # 'r-' Red-Color

plt.subplot(1,2,2)  # 1 -- Rows  2 -- Cols  2 -- Plot Number.
plt.plot(y,y,'g-')
```

```python

# In this example we are creating sub plots.

# Note : The plot number is common in all - so it will draw into same plots.

plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(x,y,'r-')  # 'r-' Red-Color

plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(y,x,'g-')  # 'g-' Green-Color

plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(x,x,'b-')  # 'b-' Blue-Color
```

```python
plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(x,y,'r-')  # 'r-' Red-Color
plt.title("Multi ploting Example")

plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(y,x,'g-')  # 'g-' Green-Color

plt.subplot(1,2,1)  # 1 -- Rows  2 -- Cols  1 -- Plot Number.
plt.plot(x,x,'b-')  # 'b-' Blue-Color
```

# Object-Oriented

```python
fig = plt.figure() # fig object is creating 
```

```python
# Creating object
fig = plt.figure()

# using object reference variable 'fig' calling methods of figure.
axes = fig.add_axes((0.1,0.1,0.8,0.8))

# This will show an empty figure 
```

```python
# Creating object
fig = plt.figure()

# using object reference variable 'fig' calling methods of figure.
axes = fig.add_axes((0.1,0.1,0.8,0.8)) # (left, bottom, width, height)

# Adding Co-ordinates to Plot
axes.plot(x,y)
axes.set_xlabel('X - Label')
axes.set_ylabel('Y - Label')
axes.set_title("Object")
```

```python
# Creating object
fig = plt.figure()

 # (left, bottom, width, height) --> Note these must be b/w 0 and 1 only.
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_title("Outer")
axes1.plot(x,y,'r-')

axes2 = fig.add_axes([0.5,0.15,0.3,0.3])
axes2.set_title("Inner")
axes2.plot(x,y,'g-')
```
